```python
# Importing Pandas library for data manupulation and analysis

import pandas as pd
```

```python
# Importing the "friends_initial.csv" file
# Using the "read_csv" method to load it into a data frame

csv_file = 'friends_initial.csv'
df = pd.read_csv(csv_file)
df
```

```python
# Splitting the 'title' column into season and episode, and title i.e. different columns for episode title and episode information

df[['Season_Episode', 'Title']] = df['title'].str.split("∙", n=1, expand=True)
df
```

```python
# Dropping column

df = df.drop(columns=['title'])
df
```

```python
# Splitting the Season and Episode info into seperate columns

df[['Season', 'Episode']] = df['Season_Episode'].str.split(".", n=1, expand=True)
df
```

```python
# Dropping column

df = df.drop(columns=['Season_Episode'])
df
```

```python
# Replacing the "S" and "E" in the Season and Episode column, respectively

df['Season'] = df['Season'].str.replace(r'S', '')
df['Episode'] = df['Episode'].str.replace(r'E', '')
df
```

```python
# Splitting the 'rating' column usign regex into two columns

df[['Rating', 'No. of Votes']] = df['rating'].str.extract(r'(.*\/10)\s\((.*)\)')
df
```

```python
# Removing the '/10' in the 'Rating' column using regex

df['Rating'] = df['Rating'].str.replace(r'/10', '')
df
```

```python
# Renaming the columns (used a dictioanry to rename multiple columns)

column_mapping = {'air_date': 'Original Air Date', 'description': 'Synopsis'}
df = df.rename(columns=column_mapping)
df
```

```python
# Dropping column

df = df.drop(columns=['rating'])
df
```

```python
# Reoredring columns

reorder = ['Season', 'Episode', 'Title', 'Synopsis', 'Original Air Date', 'Rating', 'No. of Votes']
df = df[reorder]
df
```

```python
# Writing to a .csv file

final_csv_file = 'imdb_friends_clean_dataset.csv'
df.to_csv(final_csv_file, index=False)
```
